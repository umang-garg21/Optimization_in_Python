### Prob 1
Build a Sudoku solver and run it on the samples above. All are assumed to be $n^2xn^2$, When solved, print out the values as a square grid to simplify interpretation. I would use the encoding $xijv$ i denotes the row, j means column, and v is the alphabet character.

```python
def makecnf(s2, n):
    # Mapping to SAT
    n_block = math.floor(math.sqrt(n))
    cnf1= []    

    # Add clauses related to fixed entries
    for i in range(0,n):
        for j in range(0,n):
            v = s2[i][j]
            if v>0:
                cnf1.append([i*n**2 + j*n +v])


    print("Initial clauses are", cnf1, "\n")

    for i in range(0,n):
        for j in range (0,n):
            # Definedness clauses per cell
            cnf1.append([i*n**2 + j*n +v  for v in range(1,n+1)])

    # Uniqueness clause per cell
    for i in range(0,n):
        for j in range (0,n):
            for vi in range(1,n+1):
                for vj in range(vi+1,n+1):
                    cnf1.append([-(i*n**2 + j*n + vi),-(i*n**2 + j*n + vj)])


    # Definedness clause for each value through a column
    for i in range(0,n):
        for v in range(1, n+1):
                cnf1.append([i*n**2 + j*n +v  for j in range(0,n)])


    # Uniqueness clause for each value in a column
    for i in range(0,n):
            for v in range(1,n+1):
                for j1 in range (0,n):
                    for j2 in range (j1+1,n):
                        cnf1.append([-(i*n**2 + j1*n + v),-(i*n**2 + j2*n + v)])

    # Definedness clause for each value through a row
    for j in range(0,n):
        for v in range(1, n+1):
                cnf1.append([i*n**2 + j*n +v  for i in range(0,n)])


    # Uniqueness clause for each value in a row
    for j in range(0,n):
            for v in range(1,n+1):
                for i1 in range (0,n):
                    for i2 in range (i1+1,n):
                        cnf1.append([-(i1*n**2 + j*n + v),-(i2*n**2 + j*n + v)])

    # Definedness clause for each value through a block
    for r_offset in range(0,n_block):
        for c_offset in range(0,n_block):
            for v in range(1,n+1):
                cnf1.append([int(r_offset*n_block*(n**2) + c_offset*n_block*n + i*n**2 + j*n +v) for i in range(0,n_block) for j in range(0,n_block)])


    # Uniqueness clause for each value through a block
    for r_offset in range(0,n_block):
        for c_offset in range(0,n_block):
            for v in range(1,n+1):
                for i1 in range(0,n_block):
                    for j1 in range(0,n_block):
                        for i2 in range(0,n_block):
                            for j2 in range(0,n_block):
                                if i1!=i2 or j1!=j2:
                                    cnf1.append([-int(r_offset*n_block*(n**2) + c_offset*n_block*n + i1*n**2 + j1*n +v), -int(r_offset*n_block*(n**2) + c_offset*n_block*n + i2*n**2 + j2*n +v) ])



    return cnf1
```

```python
def printsudoku(s2,n):
    n_block = math.sqrt(n)
    print("\n")
    for i in range(0,n):
        line = ""
        if i%n_block == 0:
            print("---------------------")
        for j in range(0,n):
            if j>0 and j%n_block == 0:
                line += "| "
            line += str(s2[i][j])+" "
        print(line)
```

```python
def cnfsolver(cnf):
    import time
    start = time.time()
    soln = pycosat.solve(cnf)
    # print("SAT Solution is", soln)
    exec_time = time.time() - start
    return exec_time, soln
```

```python
# Decoding SAT solution

def Sudokumapper(s2, soln):
    for val in soln:
        # Row value is int(val/n**2)
        if val > 0:
            i = int((val-1)/n**2)
            j = int( int((val-1)%n**2)/n  )
            # print("i,j", i, j)
            temp = val%(n)
            if temp == 0:
                s2[i][j] = temp + 4
            else:
                s2[i][j] = temp
    return s2
```
